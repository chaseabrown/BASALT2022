```python
import minedojo
```

```python
all_programmatic_ids = minedojo.tasks.ALL_PROGRAMMATIC_TASK_IDS
all_programmatic_instructions = minedojo.tasks.ALL_PROGRAMMATIC_TASK_INSTRUCTIONS

instructions = []
for i in range(0, len(all_programmatic_ids)):
    print(all_programmatic_ids[i], ":")
    first = True
    for instruction in all_programmatic_instructions[all_programmatic_ids[i]]:
        if first:
            print(instruction)
            first = False
        else:
            instructions.append(instruction)
            print(instruction)
```

```python
with open("../assets/MineDojoInstructions.txt", 'w') as file:
    for instruction in instructions:
        file.write(instruction + "\n")
```
